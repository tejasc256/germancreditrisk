```python
#Import neccessary data and libraries
import pandas as pd
import numpy as np
data = pd.read_csv("index.csv")
```

```python
newdf=pd.DataFrame(columns=['Parameter 1','Parameter 2','Parameter 3','Parameter 4','Parameter 5'])
print(data.nunique())
col_list=data.nunique().tolist()
```

```python
k=0
for i in range(1,21):
    if col_list[i]<6:
        arr=data.iloc[:,i].value_counts().sort_index().tolist() #Generate list of occurences of every unique element
        for j in range(0,len(arr)):
            arr[j]=arr[j]/10
        while len(arr)<5 :   #Append extra NaN as needed
            arr.append(np.nan)
        newdf.loc[k]=arr
        k+=1
#newdf = newdf.iloc[newdf.isnull().sum(1).sort_values(ascending=True).index]
newdf = newdf.iloc[newdf.isnull().sum(axis=1).mul(1).argsort()]
newdf
```
