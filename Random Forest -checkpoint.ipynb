```python
import numpy
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
df = pd.read_csv('training.csv')  #Initialisation
```

```python
df.head()
```

```python
#Creating lists X and Y to input into the fit method
X = []
Y = df[df.columns[0]].tolist()
for i in range(500):
    x = df.iloc[i,2:22].tolist()
    X.append(x)  
```

```python
#X, y = make_classification(n_samples=500, n_features= 20, n_informative=2, n_redundant=0, random_state=0, shuffle=False)
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(X, Y)
```

```python
table = [0, 0, 0, 0]
df1 = pd.read_csv('test.csv')
Y = df1[df1.columns[1]].tolist()
correct = 0
wrong = 0
for i in range(500):
    output = clf.predict([X[i]])
    if(output[0] == 1 and Y[i] == 1):
        table[0] = table[0] + 1
        #correct = correct + 1
    elif(output[0] == 0 and Y[i] == 1):
        table[1] = table[1] + 1
    elif(output[0] == 1 and Y[i] == 0):
        table[2] = table[2] + 1 
    else:
        table[3] = table[3] + 1
#correct = (correct/5)
#print('Accuracy = ', correct, '%')
accuracy =  ( table[0] + table[3] ) / 500
print(accuracy*100)
print(table)
```
